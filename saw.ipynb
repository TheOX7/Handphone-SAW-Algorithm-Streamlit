```python
brands_to_filter = ['Samsung']
df = df[df['brand'].isin(brands_to_filter)].reset_index(drop=True)
```

```python
df_testing = df.copy()

# 1. Ubah kolom dimensions menjadi kolom terpisah
df_testing[['length_mm', 'width_mm', 'thickness_mm']] = df_testing['dimensions'].str.extract(r'([\d.]+) x ([\d.]+) x ([\d.]+) mm')
df_testing['length_mm'] = pd.to_numeric(df_testing['length_mm'])
df_testing['width_mm'] = pd.to_numeric(df_testing['width_mm'])
df_testing['thickness_mm'] = pd.to_numeric(df_testing['thickness_mm'])

# 2. Ambil nilai pixels dari kolom display_size
df_testing['display_size_pixels'] = df_testing['display_size'].str.extract(r'(\d+ x \d+) pixels')

# 3. Ambil value memory dari kolom internal_memory
df_testing['internal_memory_GB'] = df_testing['internal_memory'].str.extract(r'(\d+) GB')
df_testing['internal_memory_GB'] = pd.to_numeric(df_testing['internal_memory_GB'], errors='coerce')

# 4. Bersihkan kolom primary_camera & secondary_camera
df_testing['primary_camera_MP'] = df_testing['primary_camera'].str.extract(r'(\d+) MP')
df_testing['primary_camera_MP'] = pd.to_numeric(df_testing['primary_camera_MP'], errors='coerce')

df_testing['secondary_camera_MP'] = df_testing['secondary_camera'].str.extract(r'(\d+) MP')
df_testing['secondary_camera_MP'] = pd.to_numeric(df_testing['secondary_camera_MP'], errors='coerce')

# 5. Ambil value mAh dari kolom battery
df_testing['battery_mAh'] = df_testing['battery'].str.extract(r'(\d+) mAh')
df_testing['battery_mAh'] = pd.to_numeric(df_testing['battery_mAh'], errors='coerce')

# 6. Buat kolom is_removable
df_testing['is_removable'] = df_testing['battery'].str.contains('Removable').astype(int)

# Hapus kolom yang sudah tidak diperlukan
df_testing = df_testing.drop(['dimensions', 'display_size', 'internal_memory', 'primary_camera', 'secondary_camera', 'battery'], axis=1)

# Tampilkan hasil akhir
df_testing.head()
```

```python
import re

# 1. Ambil nilai persentase screen-to-body-ratio dari kolom display_resolution
df_testing['screen_to_body_ratio'] = df_testing['display_resolution'].str.extract(r'~([\d.]+)% screen-to-body ratio')
df_testing['screen_to_body_ratio'] = pd.to_numeric(df_testing['screen_to_body_ratio'], errors='coerce')

# 2. Ambil value GHz/MHz dari kolom CPU
cpu_speed_extract = df_testing['CPU'].str.extract(r'(\d+(\.\d+)?) (GHz|MHz)').astype({0: 'float'})
df_testing['cpu_speed'] = cpu_speed_extract[0] * df_testing['CPU'].apply(lambda x: 1000 if 'MHz' in x else 1)  # Convert MHz to GHz

# 3. Ambil value GB/MB dari kolom RAM
ram_size_extract = df_testing['RAM'].str.extract(r'(\d+(\.\d+)?) (GB|MB)').astype({0: 'float'})
df_testing['ram_size'] = ram_size_extract[0] * df_testing['RAM'].apply(lambda x: 1024 if 'MB' in x else 1)  # Convert MB to GB

# Hapus kolom yang sudah tidak diperlukan
df_testing = df_testing.drop(['display_resolution', 'CPU', 'RAM'], axis=1)

# Tampilkan hasil akhir
df_testing.head()
```

```python
def clean_display_type(display_type):
    # Identifikasi tipe teknologi layar utama
    tech_type = re.findall(r'(Super AMOLED|IPS LCD|PLS|TFT|AMOLED)', display_type)
    
    # Identifikasi jumlah warna yang didukung
    color_support = re.findall(r'(\d+M|\d+K) colors', display_type)
    
    # Identifikasi fitur tambahan jika ada
    additional_features = re.findall(r'(Super Flexible AMOLED|Super AMOLED Plus|SC-LCD)', display_type)
    
    return {
        'technology_type': tech_type[0] if tech_type else None,
        'color_support': color_support[0] if color_support else None,
        'additional_features': additional_features
    }

# Terapkan fungsi ke kolom 'display_type' dan buat kolom baru
df_testing['display_type_cleaned'] = df_testing['display_type'].apply(clean_display_type)

# Pisahkan kolom baru menjadi beberapa kolom terpisah
df_testing = pd.concat([df_testing, pd.json_normalize(df_testing['display_type_cleaned'])], axis=1)

# Hapus kolom 'display_type_cleaned' jika tidak diperlukan lagi
df_testing = df_testing.drop(['display_type_cleaned', 'display_type'], axis=1)

# Tampilkan hasil akhir
df_testing.head()
```
